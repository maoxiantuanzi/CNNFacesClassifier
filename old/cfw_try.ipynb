```python
import os
import skimage.data
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib
import matplotlib.pyplot as plt
import random
```

```python
def load_data(data_dir):
     # Get all subdirectories of data_dir. Each represents a label.
    directories = [d for d in os.listdir(data_dir)
                   if os.path.isdir(os.path.join(data_dir, d))]
    # Loop through the label directories and collect the data in
    # two lists, labels and images.
    labels = []
    images = []
    i=0
    for d in directories:
        label_dir = os.path.join(data_dir, d)
        # print(label_dir)
        file_names = [os.path.join(label_dir, f)
                      for f in os.listdir(label_dir) if f.endswith(".jpg")]
        # For each label, load it's images and add them to the images list.
        # And add the label number (i.e. directory name) to the labels list.
        for f in file_names:
            images.append(skimage.data.imread(f))
            labels.append(int(i))
        i+=1
    return images, labels
```

```python
ROOT_PATH = "/home/pengtt/assignment/MSRA-CFW/Sample"
data_dir = os.path.join(ROOT_PATH, "thumbnails_features_deduped_sample")
X, y= load_data(data_dir)
# print(y)
print("load_data done")
```

```python
# 避免过拟合，采用交叉验证，验证集占训练集20%，固定随机种子（random_state)
X_train,X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.2,
                                                   random_state = 0)

print("train_test_split done")
```

```python
n_train = len(X_train)
n_test = len(X_test)
n_classes = len(set(y_train))
print("Unique Labels: {0}\nTotal Images: {1}".format(n_classes, len(X)))
```

```python
def display_images_and_labels(images, labels):
    """Display the first image of each label."""
    unique_labels = set(labels)
    plt.figure(figsize=(15, 15))
    i = 1
    for label in unique_labels:
        # Pick the first image for each label.
        image = images[labels.index(label)]
        plt.subplot(8, 8, i)  # A grid of 8 rows x 8 columns
        plt.axis('off')
        plt.title("Label {0} ({1})".format(label, labels.count(label)))
        i += 1
        _ = plt.imshow(image)
    plt.show()

display_images_and_labels(X, y)
```

```python
display_images_and_labels(X_test, y_test)
```

```python
def display_label_images(images, label):
    """Display images of a specific label."""
    limit = 24  # show a max of 24 images
    plt.figure(figsize=(15, 5))
    i = 1

    start = y.index(label)
    end = start + y.count(label)
    for image in images[start:end]:
        plt.subplot(3, 8, i)  # 3 rows, 8 per row
        plt.axis('off')
        i += 1
        plt.imshow(image)
        if i>24:
            break
    plt.show()

display_label_images(X, 0)
```

```python
# Resize images
X_train32 = [skimage.transform.resize(image, (128, 128))
                for image in X_train]
X_test32 = [skimage.transform.resize(image, (128, 128))
                for image in X_test]
display_images_and_labels(X_train32, y_train)
```
